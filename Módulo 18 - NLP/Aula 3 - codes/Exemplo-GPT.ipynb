```python
from transformers import pipeline

# Carregue os modelos para tradução e sumarização
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-pt-en")
summarizer = pipeline("summarization")

# Texto em português que você deseja traduzir e resumir
texto_em_portugues = """
Seja qual for a tarefa que você deseja realizar, é importante ter um objetivo claro em mente.
O planejamento e a organização são fundamentais para alcançar o sucesso em qualquer empreendimento.
"""

# Traduza o texto para inglês
texto_em_ingles = translator(texto_em_portugues, src_lang="pt", tgt_lang="en")[0]["translation_text"]

# Resuma o texto em inglês
resumo = summarizer(texto_em_ingles, max_length=50, min_length=20, do_sample=False)[0]["summary_text"]

# Imprima o texto original em português, o texto traduzido em inglês e o resumo
print("Texto em Português:")
print(texto_em_portugues)

print("\nTexto em Inglês:")
print(texto_em_ingles)

print("\nResumo:")
print(resumo)
```
